````python
%%file main.cpp
#include <iostream>

int main(int argc, char *argv[])
{
   std::cout << "Hello CMake!" << std::endl;
   return 0;
}
````

````python
%%file CMakeLists.txt
cmake_minimum_required(VERSION 3.5) # 设置 CMake 最小版本
project (hello_cmake) # 设置项目名
add_executable(hello_cmake main.cpp) # 添加可执行文件
````

使用外部构建，创建可以位于文件系统上任何位置的构建文件夹。所有临时构建和目标文件都位于此目录中，以保持源代码树的整洁。

运行下述代码，新建 `build` 构建文件夹，并运行 `cmake` 命令：

```bash
mkdir -p build
cd build/
cmake ..
```

可以看到，`build` 文件夹下生成了许多二进制文件，如果要从头开始重新创建 `cmake` 环境，只需删除构建目录 `build`，然后重新运行 `cmake` 命令。
